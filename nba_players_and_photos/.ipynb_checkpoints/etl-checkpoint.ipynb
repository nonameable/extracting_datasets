```python
import pandas as pd
```

```python
file_path = './players_stats.csv'
players = pd.read_csv(file_path)
```

```python
# lets select some columns
cols_to_select = ['Name', 'Games Played', 'MIN', 'PTS', '3P%', 'FT%', 'OREB', 'DREB', 'AST', 'STL', 'BLK','Age',
 'Birth_Place',
 'Birthdate',
 'Collage',
 'Experience',
 'Height',
 'Pos',
 'Team',
 'Weight',
 'BMI' ]
```

```python
cols_to_select
```

```python
players_selection = players[cols_to_select]
```

```python
players_selection.head()
```

```python
names = list(players_selection['Name'])
```

```python
with open("players.txt", "w") as f: 
    for name in names:
        f.write(name + "\n")
```

```python
# we obtain the photos and save them in the players folder
from google_images_download import google_images_download
response = google_images_download.googleimagesdownload()
arguments = {"keywords_from_file":"./players.txt","limit":1,"print_urls":True,
             "format":"png", "no_directory":False, "output_directory": "players", "no_numbering":True}
#paths = response.download(arguments)   #uncomment this to download the photos
#print(paths)
```

```python
# formatting images You can only run this cell once, after that, all the names will be lowercase
import os
for name in names:
    path = "./players/" + name + "/"
    file_name = next(os.walk(path))[2][0]
    new_file_name = name.lower().replace(" ", "-")
    # rename the png file
    os.rename(path + file_name, path + new_file_name + ".png")
    # rename the folder for an easier url
    os.rename("./players/" + name,"./players/" + new_file_name )
    
    

    
    
```

```python
basic_url = "https://raw.githubusercontent.com/nonameable/extracting_datasets/master/nba_players_and_photos/players"
for name in names:
    correct_name = name.lower().replace(" ", "-")
    url = basic_url + "/" + correct_name + "/" + correct_name + ".png" 
    players_selection.loc[players_selection['Name'] == name, 'url'] = url
```
